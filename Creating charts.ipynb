```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
data_folder = 'data'
output_folder = 'output'

if not os.path.exists(data_folder):
    os.mkdir(data_folder)
if not os.path.exists(output_folder):
    os.mkdir(output_folder)
```

```python
def download(url):
    filename = os.path.join(data_folder, os.path.basename(url))
    if not os.path.exists(filename):
        from urllib.request import urlretrieve
        local, _ = urlretrieve(url, filename)
        print('Downloaded ' + local)
```

```python
# we have 12 diff csv files containing crime data for each month of 2020. download all of them to data folder
files = [
  '2020-01-metropolitan-street.csv',
  '2020-02-metropolitan-street.csv',
  '2020-03-metropolitan-street.csv',
  '2020-04-metropolitan-street.csv',
  '2020-05-metropolitan-street.csv',
  '2020-06-metropolitan-street.csv',
  '2020-07-metropolitan-street.csv',
  '2020-08-metropolitan-street.csv',
  '2020-09-metropolitan-street.csv',
  '2020-10-metropolitan-street.csv',
  '2020-11-metropolitan-street.csv',
  '2020-12-metropolitan-street.csv'
]

data_url = 'https://github.com/spatialthoughts/python-dataviz-web/raw/main/data/crime/'

for f in files:
  url = os.path.join(data_url + f)
  download(url)
```

```python
# data pre-processing
# merge 12 csv files into single dataframe. use 'pd.concat()' method to merge a list of multiple dataframes into one dataframe
df_list = []

for f in files:
  filepath = os.path.join(data_folder, f)
  df = pd.read_csv(filepath)
  df_list.append(df)

merged_df = pd.concat(df_list)
```

```python
merged_df
```

```python
# create a pie chart
# pandas 'groupby' method allows us to calculate group stats
type_counts = merged_df.groupby('Crime type').size()
type_counts
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15,10)
type_counts.plot(kind='pie', ax=ax)
plt.show()
```

```python
wedgeprops={'linewidth': 1, 'edgecolor': 'yellow'}
textprops= {'fontsize': 15, 'fontstyle': 'italic'}

fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15,10)

type_counts.plot(kind='pie', ax=ax,
                 wedgeprops=wedgeprops,
                 textprops=textprops)

ax.set_title('Crime Types', fontsize=18)
ax.set_ylabel('')
plt.tight_layout()
plt.show()
```

```python
# create a bar chart
# group the data by month
monthly_counts = merged_df.groupby('Month').size()
monthly_counts
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15,7)

monthly_counts.plot(kind='bar', ax=ax)
monthly_counts.plot(kind='line', ax=ax,
                    color='red',
                    marker='o')

ax.set_title('Total Crime by Month')
ax.set_ylabel('Total Incidents')
plt.show()
```

```python
bicycle_thefts = merged_df[merged_df['Crime type'] == 'Bicycle theft']
monthly_bicycle_thefts = bicycle_thefts.groupby('Month').size()
monthly_bicycle_thefts


fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15,7)

monthly_bicycle_thefts.plot(kind='line', ax=ax,
                    color='yellow',
                    marker='o')

ax.set_title('Crime Types')
ax.set_ylabel('bicycle thefts')
plt.show()
```
