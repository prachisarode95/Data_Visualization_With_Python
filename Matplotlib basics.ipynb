```python
import os
import matplotlib.pyplot as plt
```

```python
# matplotlib terminology
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(5, 5)
plt.show()
```

```python
# plot a single point using matplotlib. display a point at the coordinate (0.5, 0.5)
point = (0.5, 0.5)

fig, ax = plt.subplots(1, 1)
fig.set_size_inches(5,5)
ax.plot(point[0], point[1], color='red', marker='o')
plt.show()
```

```python
points = [(0.1, 0.5), (0.5, 0.5), (0.9, 0.5)]

# But to plot it, matplotlib require 2 separate lists or x and y coordinates. Here we can use the zip() function to create list of x and y coordinates.
x, y = zip(*points)
print(x)
print(y)
```

```python
fig, ax = plt.subplots(1, 1)
fig.set_size_inches(5,5)
ax.plot(x, y, color='green', marker='o', linestyle='None')

output_folder = '/content/drive/MyDrive/Colab Notebooks/python_dataviz/output'
output_path = os.path.join(output_folder, 'simple1.png')
plt.savefig(output_path)
plt.show()
```

```python
# exercise 1
point1 = [(0.2, 0.5), (0.3, 0.5), (0.4, 0.5)]

a, b = zip(*point1)

fig, ax = plt.subplots(1, 1)
fig.set_size_inches(5,5)
ax.plot(a, b, color='green', marker='^', linestyle='None')

output_folder = '/content/drive/MyDrive/Colab Notebooks/python_dataviz/output'
output_path = os.path.join(output_folder, 'simple2.png')
plt.savefig(output_path)
plt.show()
```
